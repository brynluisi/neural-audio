```python
import scipy.io
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn-whitegrid')

import torchaudio
```

```python
import os
dirname = os.path.abspath('')
rootdir = os.path.split(dirname)[0]

H1_TRAINING_INPUT_PATH = "".join([rootdir, "/data/train/ht1-input.wav"])
H1_TRAINING_TARGET_PATH = "".join([rootdir, "/data/train/ht1-target.wav"])

H1_LSTM_OUTPUT_PATH = "".join([rootdir, "/output/lstm_mlp_premp_filter_mse_h1.wav"])
```

```python
train_target, fs = torchaudio.load(H1_TRAINING_TARGET_PATH)
train_output, fs = torchaudio.load(H1_LSTM_OUTPUT_PATH)


target_data = train_target.numpy().reshape(-1)
output_data = train_output.numpy().reshape(-1)
```

```python
length = (len(target_data) - 1)/ fs
time = np.linspace(0., 0.005, 200)

fig, ax = plt.subplots(figsize=(4.5, 3), dpi=100)
ax.plot(time, target_data[int(4 * fs): int(4 * fs) + 200 ], label="Target", color='b')
ax.plot(time, output_data[int(4 * fs): int(4 * fs) + 200 ], label="Output", color='orange')
plt.title("LSTM Model Target vs Output 200 Samples")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```
